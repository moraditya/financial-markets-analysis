```python
# import relevant libraries

import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# load the combined_df 

stocks = pd.read_csv(r"C:\Users\adityamxr\Desktop\Projects\stock-market-analysis\data\combined_stock_data.csv")
```

```python
# verify import
stocks.head()
```

```python
# print dataframe info
stocks.info()
```

```python
# check for missing values

missing_values = stocks.isnull().sum()

print(missing_values)
```

```python
# drop Adj Close feature

stocks = stocks.drop(columns= 'Adj Close')
```

```python
stocks.head()
```

```python
# missing percentage in moving averages 7 and 30

missing_percentage = stocks.isnull().mean()*100
print(missing_percentage)
```

### For the first 6 days and for the first 29 days, moving averages would be null, so it's appropriate to keep those values filled.

```python
# check for nunqiue values in all columns

for column in stocks.columns:
    print(f'{column}: {stocks[column].nunique()}')
```

### All columns nunique values look appropriate.

```python
# check .info()

stocks.info()
```

```python
# convert 'date' to datetime

stocks['Date'] = pd.to_datetime(stocks['Date'])
```

```python
print(stocks['Date'].dtype)
```

```python
duplicate_rows = stocks.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_rows}")
```

### Sanity Check (ensure there aren't any negative stock prices)

```python
# Check for negative or zero stock prices
invalid_prices = stocks[stocks['Close'] <= 0]
print(f"Rows with invalid prices: {len(invalid_prices)}")
```

```python
# final check for nulls and datatypes

print(stocks.isnull().sum())
print(stocks.dtypes)
```

```python
# remove timezone information and convert to naive datetime64[ns]
stocks['Date'] = stocks['Date'].dt.tz_convert(None)

# verify the data type of the 'Date' column
print(stocks['Date'].dtype)
```
