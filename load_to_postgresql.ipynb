```python
# import relevant libraries

import os
import pandas as pd
from sqlalchemy import create_engine
```

```python
# import cleaned_stock_df 

cleaned_data = pd.read_csv(r"C:\Users\adityamxr\Desktop\Projects\stock-market-analysis\data\cleaned_stock_df.csv")
```

```python
# verify import
cleaned_data.head()
```

```python
# fetch PostgreSQL credentials from environment variables
postgres_user = os.getenv('POSTGRES_USER')
postgres_pass = os.getenv('POSTGRES_PASSWORD')
postgres_db = os.getenv('POSTGRES_DB')
postgres_host = os.getenv('POSTGRES_HOST')
postgres_port = os.getenv('POSTGRES_PORT')
```

```python
# create a connection to PostgreSQL using the environment variables
engine = create_engine(f'postgresql://{postgres_user}:{postgres_pass}@{postgres_host}:{postgres_port}/{postgres_db}')
```

```python
# load the cleaned dataframe into postgresql
cleaned_data.to_sql('stock_data', engine, if_exists='replace', index =False)
```
